```python
#dropping colmns
titanic.drop(columns=["boat","body"], inplace=True)
```

```python
titanic["age"].info()
```

```python
#converting age to numeric
titanic["age"]=pd.to_numeric(titanic["age"], errors="coerce")
```

```python
titanic
```

```python
#converting symbols values to nan value from cabin column
for i in titanic["cabin"]:
    if i == "?":
        titanic["cabin"] = titanic["cabin"].replace("?",np.nan)
```

```python
titanic
```

```python
#converting symbols to nan value from home.dest column
for i in titanic["home.dest"]:
    if i == "?":
        titanic["home.dest"] = titanic["home.dest"].replace("?",np.nan)
```

```python
titanic
```

```python
titanic.isna().sum()
```

```python
#adding new column to show total number of relatives
titanic["relatives"] = titanic["sibsp"] + titanic["parch"]
```

```python
titanic.insert(7,"total relatives",titanic["relatives"])
```

```python
titanic.age.value_counts(dropna=False)
```

```python
Children = titanic["age"] < 13
c_value = titanic[Children]["survived"].value_counts().sort_values()
print("\nChildren survived: \n",c_value)


Teenagers = titanic["age"].between(13,17)
t_value = titanic[Teenagers]["survived"].value_counts()
print("\nTeenages survived: \n",t_value)

Adults = titanic["age"] >= 18 & (titanic["age"] < 60)
a_value = titanic[Adults]["survived"].value_counts()
print("\nAdults survived: \n ",a_value)

Older_adults = titanic["age"] >= 60
o_value = titanic[Older_adults]["survived"].value_counts()
print("\nSenior citizens:\n ", o_value)
```

```python
#displays the total percentage of survived and death from each category
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.title("Children :")
plt.pie(c_value, autopct="%1.1f", labels=["not survived"," survived"]) 

plt.subplot(2,2,2)
plt.title("Teenagers")
plt.pie(t_value, autopct="%1.1f", labels=["not survived","survived"]) 

plt.subplot(2,2,3)
plt.title("Adults")
plt.pie(a_value, autopct="%1.1f", labels=["not survived","survived"])


plt.subplot(2,2,4)
plt.title("Senior Citizens")
plt.pie(o_value, autopct="%1.1f", labels=["not survived","survived"])

plt.show()
```
